# Movielens Dataset

```python
movies = pd.read_parquet('../data/movielens.parquet')
```

```python
movies.head()
```

```python
movie_explorer = EventExplorer(
    event_data=movies,
    event_name_col='rating',
    event_time_col='timestamp',
    coarse_grain=timedelta(days=365),
    fine_grain='d',
)
```

```python
movie_explorer.plot_event_volume()
```

```python
movies['title'].value_counts().head()
```

```python
movie_explorer.explore_entity(entity='title', item='Matrix, The (1999)')
```

```python
movie_explorer.explore_entity(entity='genres', item='Drama')
```

```python
movie_explorer.event_name_col = 'primary_genre'
```

```python
# TODO: add "other" category
```

```python
fig, df_to_plot = movie_explorer.make_sankey(
    entity='userId',
    first_event='Drama',
    last_event=None,
    n_events=5,
    session_length=timedelta(days=360*30),
    include_duplicates=True,
    ignore_list=[],
)

fig.show()
```
